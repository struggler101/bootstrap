```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("DATA.csv")
df
```

```python
%matplotlib inline
plt.xlabel("AREA(sqr ft)");
plt.ylabel("PRICE(US$)");
plt.scatter(df.Area,df.Price,color="red",marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
m = reg.coef_
x = 3300
c = reg.intercept_
y = m*x+c
y
```

```python
d=pd.read_csv("eg.csv");
d
```

```python
reg.predict(d)
```

```python
p = reg.predict(d)
d['prices'] = p
```
